```python
with open('num.csv', 'w') as fi:

    for i in range(9):
        a = tuple(range(i+1,i+6))
#         a = str(a)[1:-1]
        fi.write("%s, %s, %s, %s, %s\n"%a)
```

读入csv文件并解析

```python
with open('num.csv', 'r') as fi:
    data = fi.read()
alist = []

new_data = data.splitlines()
for i in new_data:
    a =i.split(', ')
    a =[int(i) for i in a]
    alist.append(a)
print(alist)
        
```

对csv数据进行处理

```python
import numpy as np
data_np = np.array(alist,dtype=np.float)
def describe(pre_data):
    max_data = np.max(pre_data)
    min_data = np.max(pre_data)
    mean_data = np.mean(pre_data)
    median_data = np.median(pre_data)
    des_dict = {'max_data':max_data ,'min_data':min_data,'mean_data':mean_data,'median_data':median_data}
    return des_dict
```

```python
rows, cols = data_np.shape
for i in range(rows):
    print(describe(data_np[i,:]),data_np.dtype)
```
